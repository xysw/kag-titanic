# EDA
- PassengerId has no effect on target - only each row number
- Trying to predict "Survived"
- Pclass is socio-economic class
- name, sex, age are self-explanatory
- sibsp and parch are number of siblings and spouse, and parents and children
- ticket is ticket number, fare is passenger fare, cabin is cabin number
- embarked is port of embarkation and has 3 values - C, Q, S

observations:
- missing data in age, cabin, and embarked in training 
- missing data in age, cabin, and fare in test
- around 22-23% of cabin is not missing in both, the missing 77-78% would be hard to fill

decisions:
- fill in age with median age by sex and class
- fill in embarked with research
- fill in fare with median fare for class, sibsp, parch
- make cabin category one letter, fill in cabin with M for missing
- dropped passengerId because useless

```python
train_data_corr = train_data.corr(numeric_only=True).abs().unstack().sort_values(kind="quicksort", ascending=False).reset_index()
train_data_corr.rename(columns={"level_0": "Feature 1", "level_1": "Feature 2", 0: "Correlation Coefficient"},
                    inplace=True)
train_data_corr[train_data_corr["Feature 1"] == "Age"]
```

Age and Pclass have the highest correlation coefficient. We can be more accurate by also factoring in the sex along with Pclass.

```python
# only in train data:
age_by_pclass_sex = train_data.groupby(['Sex', 'Pclass'])['Age'].median()

for i in range(1, 4):
    for sex in ['female', 'male']:
        print("Median age of Pclass {} {}s: {}".format(i, sex, age_by_pclass_sex[sex][i]))
print(f"Median age of all passengers in train data: {train_data['Age'].median()}")
```

We fill in the missing values in age with the median ages of that person's sex and pclass group - for both training and test sets

```python
train_group = train_data.groupby(['Sex', 'Pclass'])['Age'].transform('median')
train_data['Age'].fillna(train_group, inplace=True)

test_group = test_data.groupby(['Sex', 'Pclass'])['Age'].transform('median')
test_data['Age'].fillna(test_group, inplace=True)
```

now focus on missing embarked values. 2 are missing from training set, none are missing from test set

```python
train_data[train_data['Embarked'].isnull()]
```

both passengers are female and have same class, same cabin, same fare, and same ticket number => they probably know each other.
googling stone, mrs. george nelson reveals that seh embarked from S (Southampton) with her maid Amelie Icard. we can fill in this info.

```python
train_data['Embarked'] = train_data['Embarked'].fillna('S')
```

missing 1 value for fare in test set:

```python
test_data[test_data['Fare'].isnull()]
```

he is male, 3rd class, 0 sibsp and 0 parch - we will fill this value with the median for third class, no sibsp, and no parch. (we can assume fare and sibsp and parch are related)

```python
median_fare = test_data.groupby(['Pclass', 'Parch', 'SibSp']).Fare.median()[3][0][0]
test_data['Fare'] = test_data['Fare'].fillna(median_fare)
```

the first letter of the cabin values are the decks where the cabins are located. we create a new feature "decks" with just the first letter and we replace missing cabins with 'M' for missing.

```python
train_data['Deck'] = train_data['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else 'M')

test_data['Deck'] = test_data['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else 'M')
```

in the dataset, there is only one person with T cabin and he is first class (no one past T). \
we group him with A cabin, which is also a first class cabin

```python
t_to_a = train_data[train_data['Deck'] == 'T'].index
train_data.loc[t_to_a, 'Deck'] = 'A'
test_data.loc[t_to_a, 'Deck'] = 'A'
```

```python
# plotting survival rates of all decks:
# (from user on kaggle)

train_data_decks_survived = train_data.groupby(['Deck', 'Survived']).count().drop(columns=['Sex', 'Age', 'SibSp', 'Parch', 'Fare', 
                                                                                   'Embarked', 'Pclass', 'Cabin', 'PassengerId', 'Ticket']).rename(columns={'Name':'Count'}).transpose()

def get_survived_dist(df):
    
    # Creating a dictionary for every survival count in every deck
    surv_counts = {'A':{}, 'B':{}, 'C':{}, 'D':{}, 'E':{}, 'F':{}, 'G':{}, 'M':{}}
    decks = df.columns.levels[0]    

    for deck in decks:
        for survive in range(0, 2):
            surv_counts[deck][survive] = df[deck][survive][0]
            
    df_surv = pd.DataFrame(surv_counts)
    surv_percentages = {}

    for col in df_surv.columns:
        surv_percentages[col] = [(count / df_surv[col].sum()) * 100 for count in df_surv[col]]
        
    return surv_counts, surv_percentages

def display_surv_dist(percentages):
    
    df_survived_percentages = pd.DataFrame(percentages).transpose()
    deck_names = ('A', 'B', 'C', 'D', 'E', 'F', 'G', 'M')
    bar_count = np.arange(len(deck_names))  
    bar_width = 0.85    

    not_survived = df_survived_percentages[0]
    survived = df_survived_percentages[1]
    
    plt.figure(figsize=(20, 10))
    plt.bar(bar_count, not_survived, color='#b5ffb9', edgecolor='white', width=bar_width, label="Not Survived")
    plt.bar(bar_count, survived, bottom=not_survived, color='#f9bc86', edgecolor='white', width=bar_width, label="Survived")
 
    plt.xlabel('Deck', size=15, labelpad=20)
    plt.ylabel('Survival Percentage', size=15, labelpad=20)
    plt.xticks(bar_count, deck_names)    
    plt.tick_params(axis='x', labelsize=15)
    plt.tick_params(axis='y', labelsize=15)
    
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), prop={'size': 15})
    plt.title('Survival Percentage in Decks', size=18, y=1.05)
    
    plt.show()

all_surv_count, all_surv_per = get_survived_dist(train_data_decks_survived)
display_surv_dist(all_surv_per)
```

# Feature Engineering:
- binned continuous features:\
    fare - 12 quantile based bins chosen\
    age - 10 quantile based bins chosen
- created family size feature by adding sibsp, parch, and 1 (self), and grouped into 4 categories
- frequency encoded ticket_number feature
- extracted titles e.g. Mr, Mrs, Miss from the "Name" feature and grouped into 5 categories
- dropped features we don't need / already changed
- label encoded non-numerical features: embarked, sex, deck, title, family_size_grouped, age, fare
- one hot encoded categorical features: embarked, sex, deck, title, family_size_grouped, pclass
- scaled the data with standard scaler

```python
train_data['Fare'] = pd.qcut(train_data['Fare'], 12)
test_data['Fare'] = pd.qcut(test_data['Fare'], 12)
```

```python
train_data['Age'] = pd.qcut(train_data['Age'], 10)
test_data['Age'] = pd.qcut(test_data['Age'], 10)
```

```python
train_data
```

making family_size_grouped feature and ticket_frequency feature\
ticket frequency differs to family because people could've travelled with friends or servants\
ticket frequency is not grouped like family size because then they are very similar / no additional info gain

```python
train_data['Family_Size'] = train_data['SibSp'] + train_data['Parch'] + 1
test_data['Family_Size'] = test_data['SibSp'] + test_data['Parch'] + 1

family_map = {1: 'Alone', 2: 'Small', 3: 'Small', 4: 'Small', 5: 'Medium', 6: 'Medium', 7: 'Large', 8: 'Large', 11: 'Large'}
train_data['Family_Size_Grouped'] = train_data['Family_Size'].map(family_map)
test_data['Family_Size_Grouped'] = test_data['Family_Size'].map(family_map)
```

```python
train_data['Ticket_Frequency'] = train_data.groupby('Ticket')['Ticket'].transform('count')
test_data['Ticket_Frequency'] = test_data.groupby('Ticket')['Ticket'].transform('count')
```

titles were extracted from names and grouped into 5 features - Mr, Mrs, Miss/Ms, Master, Dr/Military/Noble/Clergy

```python
train_data['Title'] = train_data['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
test_data['Title'] = test_data['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
# barplots from user on kaggle
fig, axs = plt.subplots(nrows=2, figsize=(20, 20))
sns.barplot(x=train_data['Title'].value_counts().index, y=train_data['Title'].value_counts().values, ax=axs[0])

axs[0].tick_params(axis='x', labelsize=10)
axs[1].tick_params(axis='x', labelsize=15)

for i in range(2):    
    axs[i].tick_params(axis='y', labelsize=15)

axs[0].set_title('Title Feature Value Counts', size=20, y=1.05)

# --- grouping ---

train_data['Title'] = train_data['Title'].replace(['Miss','Ms', 'Mlle', 'Lady', 'Mme', 'the Countess', 'Dona'],
                                                  'Miss/Ms')
train_data['Title'] = train_data['Title'].replace(['Dr', 'Col', 'Major', 'Jonkheer', 'Capt', 'Sir', 'Don', 'Rev'],
                                          'Dr/Military/Noble/Clergy')
test_data['Title'] = test_data['Title'].replace(['Miss','Ms', 'Mlle', 'Lady', 'Mme', 'the Countess', 'Dona'],
                                                  'Miss/Ms')
test_data['Title'] = test_data['Title'].replace(['Dr', 'Col', 'Major', 'Jonkheer', 'Capt', 'Sir', 'Don', 'Rev'],
                                          'Dr/Military/Noble/Clergy')

sns.barplot(x=train_data['Title'].value_counts().index, y=train_data['Title'].value_counts().values, ax=axs[1])
axs[1].set_title('Title Feature Value Counts After Grouping', size=20, y=1.05)

plt.show()
```

label encoding and one hot encoding features (age and fare are ordinal so they are not one-hot encoded)

```python
features_non_num = ['Embarked', 'Sex', 'Deck', 'Title', 'Family_Size_Grouped', 'Age', 'Fare']

for feature in features_non_num:
    train_data[feature] = LabelEncoder().fit_transform(train_data[feature])
    test_data[feature] = LabelEncoder().fit_transform(test_data[feature])
```

```python
cat_variables = ['Pclass',
                 'Sex',
                 'Deck',
                'Embarked',
                'Title',
                'Family_Size_Grouped']

train_data = pd.get_dummies(data = train_data,
                        prefix = cat_variables,
                        columns = cat_variables,
                        dtype = int)

test_data = pd.get_dummies(data = test_data,
                        prefix = cat_variables,
                        columns = cat_variables,
                        dtype = int)
```

```python
dropped_columns = ['Family_Size','Name', 'SibSp', 'Parch', 'Ticket']

train_data.drop(columns=dropped_columns, inplace=True, axis=1)
test_data.drop(columns=dropped_columns, inplace=True, axis=1)
```

# Model
- scaled X_train

```python
train_data_no_survived = train_data.drop('Survived', axis=1, inplace=False)
X_train = StandardScaler().fit_transform(train_data_no_survived)
y_train = train_data['Survived'].values
X_test = StandardScaler().fit_transform(test_data)

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

```python
features = [x for x in train_data.columns if x not in 'Survived']
len(features)
```

```python
train_X, val_X, train_y, val_y = train_test_split(X_train, y_train, train_size = 0.8, random_state = RANDOM_STATE)
print(f'train samples: {len(train_X)}')
print(f'validation samples: {len(val_X)}')
print(f'targer proportion: {sum(train_y)/len(train_y):.4f}')
```

```python
# model building

space = {
    'learning_rate': hp.loguniform("learning_rate", np.log(0.01), np.log(1)),
    'max_depth': hp.quniform("max_depth", 3, 20, 1),
    'n_estimators': hp.quniform('n_estimators', 10, 30, 1),
    'gamma': hp.uniform ('gamma', 1,9),
    'reg_lambda' : hp.uniform('reg_lambda', 0.0, 1.0),
    'min_child_weight' : hp.quniform('min_child_weight', 0, 10, 1),
    'seed': 0
}
```

```python
def objective(space):
    model=XGBClassifier(learning_rate = space['learning_rate'],
                        max_depth = int(space['max_depth']),
                        n_estimators = int(space['n_estimators']),
                        gamma = space['gamma'],
                        reg_lambda = int(space['reg_lambda']),
                        min_child_weight = int(space['min_child_weight']),
                        eval_metric=['error', 'auc'],
                        early_stopping_rounds=10,
                        objective="binary:logistic")
    
    evaluation = [(train_X, train_y), (val_X, val_y)]

    model.fit(train_X, train_y,
                eval_set=evaluation,
                verbose=False)

    pred = model.predict(val_X)
    accuracy = accuracy_score(val_y, pred>0.5)
    print ("SCORE:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }
    
```

```python
trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 100,
                        trials = trials)
```

```python
print(f"the best hyperparameters are: \n {best_hyperparams}")
```

Found the optimal hyperparameters, now running them with a final model

```python
final_model = XGBClassifier(gamma = 1.20,
                            learning_rate = 0.59,
                            max_depth = 11,
                            min_child_weight = 6,
                            n_estimators = 22,
                            reg_lambda = 0.30,
                            verbosity = 1,
                            random_state = RANDOM_STATE,
                            early_stopping_rounds = 10)
final_model.fit(train_X,train_y, eval_set = [(val_X,val_y)])
```

```python
final_model.best_iteration
```

```python
print(f"Metrics train:\n\tAccuracy score: {accuracy_score(final_model.predict(train_X),train_y):.4f}\nMetrics test:\n\tAccuracy score: {accuracy_score(final_model.predict(val_X),val_y):.4f}")
```
